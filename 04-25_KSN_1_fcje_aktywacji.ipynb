```python
def relu(x):
  return max(0,x)
```

```python
x = np.arange(-10,10,0.1)
y = [relu(i) for i in x]
```

```python
for i in np.arange(-10,10,0.5):
  print(relu(i))
```

```python
plt.plot(x, y)
```

```python
def sigmoid(x):
  return 1. / (1+np.exp(-x))
```

```python
x = np.arange(-10,10,0.1)
y = [sigmoid(i) for i in x]
```

```python
plt.plot(x, y)
```

```python
# SOFTMAX
```

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=0)
```

```python
data = np.random.randn(3, 5)
data
```

```python
np.exp(data)
```

```python
np.sum(np.exp(data), axis=0)
```

```python
results = softmax(data)
results
```

```python

np.sum(results, axis=0)
```

```python
def tanh(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
x = np.arange(-3,3,0.1)
y = [tanh(i) for i in x]
```

```python
plt.plot(x, y)
```
