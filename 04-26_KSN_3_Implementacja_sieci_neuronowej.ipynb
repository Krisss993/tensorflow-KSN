```python

def initialize_parameters(n_x, n_h, n_y):
    W1 = np.random.rand(n_h, n_x)
    W2 = np.random.rand(n_h, n_y)
    return W1, W2
```

```python

def predict(X, W1, W2):
    _, y_pred = forward_propagation(X, W1, W2)
    return y_pred[0]
```

```python

def backpropagation(X, W1, W2, learning_rate, iters=100):

    H, y_pred = forward_propagation(X, W1, W2)
    train_loss = []

    for i in range(iters):
        error = (y_pred-y_true)[0,0]
        W2 = W2 - learning_rate * error * H.T
        W1 = W1 - learning_rate * error * np.dot(X.T, W2.T)

        _, y_pred = forward_propagation(X, W1, W2)


        print(f'Iter #{i}: y_pred {y_pred[0][0]}: loss: {abs(error)}')
        train_loss.append(abs(error))


    return W1, W2, train_loss
```

```python

def build_model():

    W1, W2 = initialize_parameters(2, 2, 1)

    W1, W2, train_loss = backpropagation(X, W1, W2, 0.01)

    model = {'W1': W1, 'W2': W2, 'train_loss': train_loss}

    return model
```

```python

model = build_model()
```

```python
loss = pd.DataFrame({'train_loss': model['train_loss']})
loss = loss.reset_index().rename(columns={'index': 'iter'})
loss['iter'] += 1
loss.head()
```
